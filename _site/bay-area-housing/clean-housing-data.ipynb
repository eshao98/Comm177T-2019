```python
import pandas as pd
import csv
```

```python
zips = pd.read_csv("test-data.csv", sep=",")
```

```python
# Split on " - ", keep first split as zip-code and second as city
split = zips["zip-code"].str.split(" - ")
zips["zip-code"] = split.str[0]
zips["city"] = split.str[-1]
```

```python
zips = zips[["zip-code", "city", "county", "2018-county-housing-wage"]]
```

```python
zips.to_csv("zip-code-city.csv", sep=",", index=False)
```

```python
zips.head()
```

```python
rhna = pd.read_csv("rhna-data.csv")
```

```python
rhna.head()
```

```python
merged = pd.merge(zips, rhna, left_on='city', right_on='jurisdiction')
merged.to_csv("data.csv", sep=",", index=False)
```

```python
merged_left = pd.merge(left=rhna, right=zips, how='left', left_on='jurisdiction', right_on='city')
merged_left.to_csv("full-data.csv", sep=",", index=False)
```
