```python
hw = pd.read_csv("hwdata.csv", sep=",")
hw
```

```python
hw = hw.groupby(['city','year']).mean().drop(['zip','onebr','twobr','threebr','fourbr'], axis=1).round(decimals=2)
```

```python
temp = pd.read_csv("temp-data.csv")
```

```python
onehw = temp.pivot(index='year', columns='city', values='onehw')
onehw['type'] = 'onehw'
twohw = temp.pivot(index='year', columns='city', values='onehw')
twohw['type'] = 'twohw'
threehw = temp.pivot(index='year', columns='city', values='onehw')
threehw['type'] = 'threehw'
fourhw = temp.pivot(index='year', columns='city', values='onehw')
fourhw['type'] = 'fourhw'
```
